```python
import os
import shutil

# Paths to folders
labels_folder = r"D:\Akash\Work\AI\Datasets\test_data\mix_test_data\mix_test\labels" # Replace with the correct path to 'labels'
new_labels_folder = r"D:\Akash\Work\AI\Datasets\test_data\mix_test_data\mix_test\new_labels_v360"  # Replace with the correct path to 'new_labels'
new_tags_folder = r"D:\Akash\Work\AI\Datasets\test_data\mix_test_data\mix_test\new_tags_v360"  # Replace with the correct path to 'new_tags'

# Ensure the output folder exists
os.makedirs(new_tags_folder, exist_ok=True)

# Iterate through files in the labels folder
for label_file in os.listdir(labels_folder):
    labels_path = os.path.join(labels_folder, label_file)
    new_labels_path = os.path.join(new_labels_folder, label_file)

    # Check if corresponding file exists in new_labels
    if os.path.isfile(new_labels_path):
        # Count lines in both files
        with open(labels_path, 'r') as lf:
            labels_lines = lf.readlines()
        with open(new_labels_path, 'r') as nlf:
            new_labels_lines = nlf.readlines()

        # Compare the number of lines
        # if len(labels_lines) != len(new_labels_lines):
        if labels_lines != new_labels_lines:
            # Copy the file to new_tags folder if the line count differs
            shutil.copy2(labels_path, os.path.join(new_tags_folder, label_file))

print("Copy operation completed.")
```

```python
import os
import shutil
import math

# Paths
images_folder = r"D:\Akash\Work\AI\Datasets\test_data\mix_test_data\mix_test\bgs"  # Replace with the correct path to your images folder
output_root_folder = r"D:\Akash\Work\AI\Datasets\test_data\mix_test_data\mix_test\grp" # Replace with the correct root output folder

# Get list of existing subfolders in output_root_folder
subfolders = [os.path.join(output_root_folder, name) for name in os.listdir(output_root_folder) if os.path.isdir(os.path.join(output_root_folder, name))]

if len(subfolders) != 7:
    raise ValueError("Exactly 7 subfolders are required in the output root folder.")

# Get sorted list of image files
image_files = sorted([f for f in os.listdir(images_folder) if os.path.isfile(os.path.join(images_folder, f))])
total_images = len(image_files)
images_per_folder = total_images // len(subfolders)
remaining_images = total_images % len(subfolders)

# Distribute images sequentially across existing folders
start_idx = 0
for i, subfolder in enumerate(subfolders):
    end_idx = start_idx + images_per_folder + (1 if i < remaining_images else 0)
    for image_file in image_files[start_idx:end_idx]:
        src_path = os.path.join(images_folder, image_file)
        shutil.copy2(src_path, subfolder)
    start_idx = end_idx

print("Image splitting completed.")


```

```python
import pandas as pd
from openpyxl import load_workbook
from openpyxl.styles import Alignment, Font, Border, Side
from openpyxl.utils import get_column_letter

def divide_number_with_table(names, total):
    base = total // len(names)
    remainder = total % len(names)

    start = 1
    distribution = []
    checkers = names[-1:] + names[:-1]  # Assign checkers by shifting the list

    for i, name in enumerate(names):
        share = base + (1 if i < remainder else 0)
        end = start + share - 1
        distribution.append([
            f"{start}-{end}",  # 
            name,  # 
            f"_/{share}",  # 
            "",  # 
            checkers[i],  # 
            f"_/{share}",  # 
            "",  # 
            ""  # 
        ])
        start = end + 1

    return distribution

# List of names
names = ["", "", "", "", "", "", ""]

# Total number to divide
N = 4921

# Get the table data
table_data = divide_number_with_table(names, N)

# Define headers
headers = ["", "", "", "", "", "", "", "\n"]

# Create DataFrame and save to Excel
output_path = "output.xlsx"
df = pd.DataFrame(table_data, columns=headers)
df.to_excel(output_path, index=False)

# Load the workbook for formatting
wb = load_workbook(output_path)
ws = wb.active

# Set column widths for better readability
col_widths = [12, 12, 10, 10, 12, 10, 10, 15]  # Adjusted for spacing
for i, width in enumerate(col_widths, start=1):
    ws.column_dimensions[get_column_letter(i)].width = width

# Style settings
bold_font = Font(bold=True)
center_align = Alignment(horizontal="center", vertical="center")
thin_border = Border(
    left=Side(style="thin"), right=Side(style="thin"),
    top=Side(style="thin"), bottom=Side(style="thin")
)

# Apply styles to header
for cell in ws[1]:  
    cell.font = bold_font
    cell.alignment = center_align
    cell.border = thin_border

# Apply styles to data cells
for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=1, max_col=ws.max_column):
    for cell in row:
        cell.alignment = center_align
        cell.border = thin_border

# Save the formatted file
wb.save(output_path)
print(f"Formatted Excel file saved as: {output_path}")
```
