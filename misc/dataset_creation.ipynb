# Random Split (train- Val)

```python
import os
import random
import shutil
from pathlib import Path

# Define the root folder containing 'images' and 'labels'
root_folder = Path(r"D:\Akash\Work\AI\tc7\2024_11_tc7_dataset_11\STILL360БЛexperience_The_new_RX_20_ld_0.94")
images_folder = root_folder / "images"
labels_folder = root_folder / "labels"

# Create train and val directories for images and labels
output_folders = {
    "train": {
        "images": root_folder / "images/train",
        "labels": root_folder / "labels/train",
    },
    "val": {
        "images": root_folder / "images/val",
        "labels": root_folder / "labels/val",
    }
}

# Ensure directories exist
for dataset, dirs in output_folders.items():
    for dir_path in dirs.values():
        dir_path.mkdir(parents=True, exist_ok=True)

# List all image files
image_files = list(images_folder.glob("*.jpg"))  # Adjust for different formats if needed
random.shuffle(image_files)  # Shuffle for random splitting

# Calculate the exact 9:1 split sizes
total_images = len(image_files)
train_count = round(total_images * 0.9)  # 90% for training
val_count = total_images - train_count  # Remaining for validation

# Assign files to train and val
train_files = image_files[:train_count]
val_files = image_files[train_count:]

# Helper function to copy files
def move_files(image_files, dataset_type):
    for image_file in image_files:
        # Determine corresponding label file
        label_file = labels_folder / f"{image_file.stem}.txt"
        
        # Skip if label file doesn't exist
        if not label_file.exists():
            print(f"Label file not found for {image_file}. Skipping...")
            continue
        
        # Copy files to respective folders
        shutil.move(image_file, output_folders[dataset_type]["images"] / image_file.name)
        shutil.move(label_file, output_folders[dataset_type]["labels"] / label_file.name)

# Copy train and val files
move_files(train_files, "train")
move_files(val_files, "val")

print(f"Dataset split completed!")
print(f"Total: {total_images}, Train: {len(train_files)}, Val: {len(val_files)}")
```

# split multiple folders and combine

```python
import os
import random
import shutil
from pathlib import Path

# Define the root folder containing subdirectories with 'images' and 'labels'
root_folder = Path(r"D:\Akash\Work\AI\2025\TC03\2025_TC2_DataPool_0.94_with_annotations")

# Get subdirectories inside the root folder
subdirectories = [d for d in root_folder.iterdir() if d.is_dir()]

# Create train and val directories for images and labels in each subdirectory
output_folders = {
    "train": {
        "images": None,  # Will set later
        "labels": None,  # Will set later
    },
    "val": {
        "images": None,  # Will set later
        "labels": None,  # Will set later
    }
}

# Create master output directories
master_output_folder = root_folder / "master_dataset"
master_output_folders = {
    "train": {
        "images": master_output_folder / "images/train",
        "labels": master_output_folder / "labels/train",
    },
    "val": {
        "images": master_output_folder / "images/val",
        "labels": master_output_folder / "labels/val",
    }
}

# Ensure master dataset directories exist
for dataset, dirs in master_output_folders.items():
    for dir_path in dirs.values():
        dir_path.mkdir(parents=True, exist_ok=True)

# Helper function to move files
def move_files(image_files, label_files, dataset_type, subdirectory, is_master=False):
    for image_file in image_files:
        # Determine corresponding label file
        label_file = label_files / f"{image_file.stem}.txt"
        
        # Skip if label file doesn't exist
        if not label_file.exists():
            print(f"Label file not found for {image_file}. Skipping...")
            continue
        
        # Create necessary output directories for this subdirectory (only for non-master)
        if not is_master:
            output_folders["train"]["images"] = subdirectory / "images/train"
            output_folders["train"]["labels"] = subdirectory / "labels/train"
            output_folders["val"]["images"] = subdirectory / "images/val"
            output_folders["val"]["labels"] = subdirectory / "labels/val"
        
        # Ensure directories exist (for subdirectories)
        if not is_master:
            for dataset, dirs in output_folders.items():
                for dir_path in dirs.values():
                    dir_path.mkdir(parents=True, exist_ok=True)
        
        # Copy or move files to respective folders
        if is_master:
            shutil.copy(image_file, master_output_folders[dataset_type]["images"] / image_file.name)
            shutil.copy(label_file, master_output_folders[dataset_type]["labels"] / label_file.name)
        else:
            shutil.move(image_file, output_folders[dataset_type]["images"] / image_file.name)
            shutil.move(label_file, output_folders[dataset_type]["labels"] / label_file.name)

# Process each subdirectory
for subdirectory in subdirectories:
    images_folder = subdirectory / "images"
    labels_folder = subdirectory / "labels"

    # List all image files
    image_files = list(images_folder.glob("*.jpg"))  # Adjust for different formats if needed
    random.shuffle(image_files)  # Shuffle for random splitting

    # Calculate the exact 9:1 split sizes
    total_images = len(image_files)
    train_count = round(total_images * 0.9)  # 90% for training
    val_count = total_images - train_count  # Remaining for validation

    # Assign files to train and val
    train_files = image_files[:train_count]
    val_files = image_files[train_count:]

    # Move files to train and val for subdirectory-specific dataset
    move_files(train_files, labels_folder, "train", subdirectory)
    move_files(val_files, labels_folder, "val", subdirectory)

    print(f"Processed subdirectory: {subdirectory.name}")
    print(f"Total: {total_images}, Train: {len(train_files)}, Val: {len(val_files)}")

# Combine all image and label files into the master dataset
for subdirectory in subdirectories:
    images_folder = subdirectory / "images"
    labels_folder = subdirectory / "labels"
    
    # List all image files in each subdirectory's train and val folders
    train_images = list(images_folder.glob("train/*.jpg"))  # Adjust for different formats if needed
    val_images = list(images_folder.glob("val/*.jpg"))
    
    # Move files to the master dataset folder
    move_files(train_images, labels_folder, "train", subdirectory, is_master=True)
    move_files(val_images, labels_folder, "val", subdirectory, is_master=True)

print("Master dataset combined and dataset split completed!")
```

# Class Check

```python
import os
from pathlib import Path

# Define the root folder containing subfolders with label files
root_folder = Path(r"D:\Akash\Work\AI\2025\TC03\OC\final\0.94\labels")

# Get all .txt label files from all subdirectories
label_files = list(root_folder.rglob("*.txt"))

# Initialize a flag for validation
all_class_0 = True

# Check each file
for label_file in label_files:
    with label_file.open("r") as file:
        lines = file.readlines()
        for line in lines:
            # Extract the class ID (the first value in each line)
            class_id = line.split()[0]
            if class_id != "0":  # If any class ID is not 0
                print(f"Non-class-0 detected in file: {label_file}")
                all_class_0 = False

# Final result
if all_class_0:
    print("All label files contain only class 0.")
else:
    print("Some label files contain classes other than 0.")
```
