```python
import os

def add_line_to_yolo_labels(label_folder, line_to_add):
    # Ensure the folder exists
    if not os.path.exists(label_folder):
        print(f"Folder '{label_folder}' does not exist.")
        return
    
    # Iterate over all txt files in the label folder
    for filename in os.listdir(label_folder):
        if filename.endswith(".txt"):
            file_path = os.path.join(label_folder, filename)
            
            # Append the new line to the file
            with open(file_path, "a") as file:
                file.write(f"{line_to_add}\n")
            print(f"Added line to {filename}")

# Example usage
label_folder = r"D:\Akash\Work\AI\2025\New folder"  # Change this to the correct path
line_to_add = "0 0.516509 0.530071 0.082547 0.112028"
add_line_to_yolo_labels(label_folder, line_to_add)
```
